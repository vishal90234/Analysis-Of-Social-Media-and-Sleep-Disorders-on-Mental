```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

# Specify the file path
file_path = r'C:\Users\DELL\Downloads\capstone 4th sem\mental_health_and_technology_usage_2022.xlsx'

# Load the dataset
data = pd.read_excel(file_path)

# Select the features and target variable
X = data[['Technology_Usage_Hours', 'Work_Environment_Impact']].copy()
y = data['Stress_Level']

# Convert 'Work_Environment_Impact' to numerical values (since it's categorical, we account for 'Neutral')
X['Work_Environment_Impact'] = X['Work_Environment_Impact'].replace({'Negative': 0, 'Neutral': 0.5, 'Positive': 1})

# Convert 'Stress_Level' to binary (assuming 'High' is the target class)
y_binary = y.apply(lambda x: 1 if x == 'High' else 0)

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.3, random_state=42)

# Initialize and fit the Logistic Regression model
logreg = LogisticRegression()
logreg.fit(X_train, y_train)

# Predict on the test set
y_pred = logreg.predict(X_test)

# Print the classification report
print(classification_report(y_test, y_pred, target_names=['Low/Medium', 'High']))
```

```python
#1.4.3 How does the combination of high social media usage and low sleep hours correlate with poor mental health status?
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = r'C:\Users\DELL\Downloads\capstone 4th sem\mental_health_and_technology_usage_2022.xlsx'
data = pd.read_excel(file_path)

# Create a new column that identifies "High Social Media Usage" and "Low Sleep Hours"
data['High_Social_Media_Usage'] = data['Social_Media_Usage_Hours'] > data['Social_Media_Usage_Hours'].median()
data['Low_Sleep_Hours'] = data['Sleep_Hours'] < data['Sleep_Hours'].median()

# Filter data for "Poor" Mental Health Status
poor_mental_health_data = data[data['Mental_Health_Status'] == 'Poor']

# Create a count plot showing poor mental health based on high social media usage and low sleep hours
plt.figure(figsize=(10, 6))
sns.countplot(data=poor_mental_health_data, x='High_Social_Media_Usage', hue='Low_Sleep_Hours', palette='coolwarm')

# Set plot labels and title
plt.xlabel('High Social Media Usage')
plt.ylabel('Count')
plt.title('High Social Media Usage and Low Sleep Hours Correlation with Poor Mental Health Status')

# Show the plot
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report

# Load the dataset
file_path = r'C:\Users\DELL\Downloads\capstone 4th sem\mental_health_and_technology_usage_2022.xlsx'
data = pd.read_excel(file_path)

# Create new columns to represent high social media usage and low sleep hours
data['High_Social_Media_Usage'] = data['Social_Media_Usage_Hours'] > data['Social_Media_Usage_Hours'].median()
data['Low_Sleep_Hours'] = data['Sleep_Hours'] < data['Sleep_Hours'].median()

# Convert 'Mental_Health_Status' to binary: 1 for 'Poor' and 0 for others
data['Mental_Health_Binary'] = data['Mental_Health_Status'].apply(lambda x: 1 if x == 'Poor' else 0)

# Define the features (High Social Media Usage and Low Sleep Hours) and the target (Mental Health Status)
X = data[['High_Social_Media_Usage', 'Low_Sleep_Hours']].astype(int)  # Convert boolean to int
y = data['Mental_Health_Binary']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and fit the SVM model
svm_model = SVC(kernel='linear')  # Using a linear kernel for this case
svm_model.fit(X_train, y_train)

# Predict on the test set
y_pred = svm_model.predict(X_test)

# Print the classification report to evaluate the model
print(classification_report(y_test, y_pred, target_names=['Not Poor', 'Poor']))
```
